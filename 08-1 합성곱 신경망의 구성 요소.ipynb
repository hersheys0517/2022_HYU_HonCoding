# 08-1 합성곱 신경망의 구성 요소
### 합성곱 신경망의 기본 개념과 동작 원리 및 풀링 계산 방법
---
<br/>
💡용어 정리💡  

❓ 합성곱 ➡️ 입력 데이터의 유용한 특성만 드러나게 하는 것  
❓ 완전 연결 신경망 ➡️ 완전 연결 층(밀집층)만 사용하여 만든 신경망 (= 밀집 신경망)  
❓ 특성 맵 ➡️ 합성곱 계산을 통해 얻은 출력  
❓ 합성곱 신경망 ➡️ 합성곱 층을 1개 이상 사용한 인공 신경망  
❓ 패딩 ➡️ 입력 배열의 주위를 가상의 원소로 채우는 것  
 --> 실제 입력값이 아니기 때문에 패딩은 0!!  
❓ 세임 패딩 ➡️ 입력과 특성 맵의 크기를 동일하게 만들기 위해 입력 주위에 0으로 패딩 하는 것  
 (for 입력의 크기 = 특성 맵의 크기)  
❓ 밸리드 패딩 ➡️ 패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만드는 경우  
 (특성 맵의 크기가 줄어들 수밖에 없음)  
❓ 스트라이드 ➡️ 커널 도장 이동의 크기  
❓ 풀링 ➡️ 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할 수행, 가중치가 없음!  
 --> 너비와 높이만 줄이고, 차원은 그대로!  
 --> 풀링의 기본 스트라이드는 2  
 (ex. (2, 2, 3) -> (1, 1, 3) 차원은 그대로)  
 1. 최대 풀링 : 최댓값(가장 큰 값)을 고름  
 2. 평균 풀링 : 평균값을 계산함

**뉴런 = 필터 = 커널**
