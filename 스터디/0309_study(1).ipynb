```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

uploaded = pd.read_csv('Tesla.csv - Tesla.csv.csv')

uploaded.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
```

```python
uploaded.value_counts('Adj Close')
```

```python
X = uploaded.drop(['Adj Close', 'Date'], axis=1)
y = uploaded['Adj Close']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

print(X_train.shape) 
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score

# 회귀 알고리즘
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

# 회귀에서는 정확도의 개념을 사용하지 않음!
rmse = (np.sqrt(mean_squared_error(y_test, y_pred))) #mse에 루트씌운거 = 평균제곱근오차
# mean_absolute_error() --> cost function 이라고 부름
mae = (mean_absolute_error(y_test, y_pred))

print("the model performance for training set")
print("----------------------")
print('RMSE is', rmse)
print('MAE is', mae)
```

```python
# 결정 트리 알고리즘
from sklearn.tree import DecisionTreeRegressor

reg = DecisionTreeRegressor()

reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

# 회귀에서는 정확도의 개념을 사용하지 않음!
rmse = (np.sqrt(mean_squared_error(y_test, y_pred))) #mse에 루트씌운거 = 평균제곱근오차
# mean_absolute_error() --> cost function 이라고 부름
mae = (mean_absolute_error(y_test, y_pred))

print("the model performance for training set")
print("----------------------")
print('RMSE is', rmse)
print('MAE is', mae)
```
