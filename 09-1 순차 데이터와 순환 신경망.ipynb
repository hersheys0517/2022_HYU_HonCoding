# 09-1 순차 데이터와 순환 신경망
### 순차 데이터의 특징 및 순환 신경망의 개념 학습
---
<br/>
💡용어 정리💡  

❓ 피드포워드 신경망 ➡️ 입력 데이터의 흐름이 앞으로만 전달되는 신경망   
 (완전 연결 신경망, 합성곱 신경망 등)  
❓ 순환 신경망 ➡️ 완전 연결 신경망 + 이전 데이터 처리 흐름을 순환하는 고리  
 - 타임스템 --> 샘플을 처리하는 한 단계  
 (순환 신경망은 이전 타임스텝의 샘플을 기억함!)  
 - 셀 --> 순환 신경망에서 부르는 층  
 - 은닉 상태 --> 셀의 출력  
 - tanh() --> 순환 신경망에서 주로 사용되는 활성화 함수
